# Time Series Analysis
##### Goal: find outliers, trends, and periodicity in the MTA turnstile data.

### Download Data:

```python
import os
import numpy as np
import pandas as pd
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as pl
from scipy import stats
from scipy.signal import argrelextrema
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_samples, silhouette_score
%matplotlib inline
```

```python
def curlfile(url, outfile):
    """Downloads file from url to PUIdata, returns output file path.
    Arguments:
    url - url location of descired file (string).
    outfile - name of downloaded file (string).
    """
    outfile = os.getenv('PUIDATA') + '/' + outfile
    os.system('curl -o ' + outfile + ' ' + url)
    return outfile
```

```python
# Curl data from github to PUIDATA
data = "https://raw.githubusercontent.com/fedhere/PUI2016_fb55/master/HW9_fb55/MTA_Fare.npy"
curlfile(data, 'MTA_Fare.npy')
```

```python
# Load data cube as numpy array.
mta_data = np.load(os.getenv('PUIDATA') + 'MTA_Fare.npy')
mta_data.shape
```

### Replace Null Values

```python
# Create empty ndarry with dtype float.
data = np.empty(mta_data.shape, dtype='float')

# Convert mta_data to float array and store them in data.
for index, arr in enumerate(mta_data):
    data[index] = arr.astype('float')
```

```python
# Replace all -1 values with nan.
for arr in data:
    arr[arr == -1] = np.nan
```

## Task 1: Event Detection
The minimum weekly ridership occured on week 127, which corresponds to date of 2012/10/26. During this week, Hurricane Sandy hit land near Brigantine, New Jersey on October 29th.

```python
# Sum over card type and stations to get total ridership by week.
cards_sum = np.nansum(data, axis=1)
cards_station = np.nansum(cards_sum, axis=0)

# Find minimum ridership index.
np.where(cards_station == min(cards_station))
```

```python
# Calculate date by adding 127 weeks to the start date.
startdate = dt.datetime.strptime('05/21/2010', '%m/%d/%Y')
deltaweek = dt.timedelta(weeks=127)
newdate = startdate + deltaweek
print newdate
```

```python
# Plot figure.
fig, (ax) = pl.subplots(nrows=1, ncols=1, figsize=(16, 4))
ax.plot(range(cards_station.shape[0]), cards_station, c='k', 
        alpha=0.6, label="MTA Ridership")
ax.plot([126, 126], [0, 40000000], c='r',
        alpha=0.6, label="Hurricane Sandy")
ax.set_xlabel('Week (Count Starting 05/21/2010)', fontsize=14, labelpad=15)
ax.set_ylabel('Weekly MTA Ridership', fontsize=14)
ax.set_title('Weekly MTA Ridership from 05/21/2010 to 02/21/2014', fontsize=16)
ax.text(0, -19000000, "Figure 1. Weekly MTA Ridership from 05/21/2010 to \
02/21/2014. The minimum event occurs at week 127 corresponding to the \
week of October \n26, 2012. Accordingly, the event that resulted in the \
reduction of MTA ridership during this week was Hurricane Sandy, which \
hit land near \nBrigantine, New Jersey on October 29th.", fontsize=14)
ax.legend(loc='lower right', ncol=1)
```

## Task 2: Ridership Type Changes

```python
# Sum of stations to get ridership by card by week.
sumstations = np.nansum(data, axis=0)
sumstations.shape
```

```python
# Load data to pandas dataframe.
cardtype = pd.DataFrame(sumstations)
```

```python
# Calculate fields for first & last 10 weeks and their ratio.
cardtype['F10_Sum'] = cardtype[range(10)].sum(axis=1)
cardtype['L10_Sum'] = cardtype[range(184, 194)].sum(axis=1)
cardtype['Difference'] = cardtype['L10_Sum'] - cardtype['F10_Sum']
cardtype['Ratio'] = cardtype['L10_Sum'] / cardtype['F10_Sum']
cardtype['Label'] = ['10 Trip', '14 Day', '1 Day', '2 Trip',
                     '30 Day', '7 Day', 'ADA', 'AFAS', 'EXP',
                     'EZPAY', 'FF', 'MC', 'MR', 'Monthly',
                     'RFM', 'RR', 'SEN', 'SPEC', 'TCMC', 'TKT',
                     'UNL', 'XBUS', 'Student']
card_data = cardtype[['F10_Sum', 'L10_Sum', 'Difference', 'Ratio', 'Label']]
card_data.sort_values('Ratio', axis=0, ascending=False)
```

The 5 card types with the largest proportional increase in use are: (1) Student Usage, (2) Mail and Ride, (3) Special, (4) Senior Citizen/Disabled, and (5) Rail Road.

The 5 card types with the largest proportional decrease in use are: (1) Easy Pay, (2) Express, (3) ADA Farecard Access System, (4) Monthly, and (5) Reduce Fare Media.

## Task 3: Periodicity

```python
# Collapse data on card type. Transpose data.
df3 = pd.DataFrame(cards_sum)
df3 = df3.T
```

```python
# Plot figure.
fig, (ax) = pl.subplots(nrows=1, ncols=1, figsize=(16, 4))
for x in df3.columns:
    ax.plot(range(194), df3[x], alpha=0.6, label="MTA Ridership")
ax.set_xlabel('Week (Count Starting 05/21/2010)', fontsize=14, labelpad=15)
ax.set_ylabel('Weekly MTA Ridership', fontsize=14)
ax.set_title('Weekly MTA Ridership at each Subway Stop from 05/21/2010 to \
02/21/2014', fontsize=16)
ax.text(0, -350000, "Figure 2. Weekly MTA ridership at each subway stop from 05/21/2010 to \
02/21/2014. This demonstrates shared trends in the ridership data \n\
ultimately aiming to inform the identification of annual periodicity.",
        fontsize=14)
```

```python
# Append fourier transform of each column to list.
ft = []
for x in df3.columns:
    ft.append(np.abs(np.fft.rfft(df3[x])))

# Append annual values (week 52) to list.
ann = []
for a in ft:
    ann.append(a[51])
```

```python
# Enumerate annual values from list.
enum = []
for tup in enumerate(ann):
    enum.append(tup)
```

```python
# Sort enumerated list by second value in tuple.
enum.sort(key=lambda tup: tup[1], reverse=True)

# Extract top 4 values for annual periodicity.
enum[0:4]
```

```python
# Calculate date by adding x weeks to the start date.
startdate = dt.datetime.strptime('05/21/2010', '%m/%d/%Y')
deltaweek = dt.timedelta(weeks=187)
newdate = startdate + deltaweek
print newdate
```

```python
# Plot figure.
fig, (ax) = pl.subplots(nrows=1, ncols=1, figsize=(16, 4))
for x in [11, 21, 32, 169]:
    ax.plot(range(194), df3[x], alpha=0.6, zorder=2, label='Station ' + str(x))

# Week with Christmass
ax.plot([31, 31], [0, 800000], alpha=0.2, color='r', label='Christmas')
ax.plot([83, 83], [0, 800000], alpha=0.2, color='r')
ax.plot([135, 135], [0, 800000], alpha=0.2, color='r')
ax.plot([187, 187], [0, 800000], alpha=0.2, color='r')

# Two weeks preceding Christmas
pl.axvspan(29, 31, color='k', alpha=0.2, zorder=1, label='Annual Periodicity')
pl.axvspan(81, 83, color='k', alpha=0.2, zorder=1)
pl.axvspan(133, 135, color='k', alpha=0.2, zorder=1)
pl.axvspan(185, 187, color='k', alpha=0.2, zorder=1)

# Labels, etc.
ax.set_xlabel('Week (Count Starting 05/21/2010)', fontsize=14, labelpad=15)
ax.set_ylabel('Weekly MTA Ridership', fontsize=14)
ax.set_title('Weekly MTA Ridership at each Subway Stop from 05/21/2010 to \
02/21/2014', fontsize=16)
ax.text(0, -350000, "Figure 2. Weekly MTA ridership of top 4 subway stops showing annual \
periodicity from 05/21/2010 to 02/21/2014. The event resulting in the \n\
annual periodicity appears to be Christmas. Accordingly, MTA could \
reasonably expect ridership to peak 2 weeks prior to Christmas.",
        fontsize=14)
ax.legend(loc='lower left')
```

## Extra Credit: Clustering

```python
# Get silhouette score for cluster range(2,6)
df4 = df3.T
range_n_clusters = range(2, 6)
for n_clusters in range_n_clusters:
    km = KMeans(n_clusters=n_clusters, random_state=324)
    cluster_labels = km.fit_predict(df4)
    silhouette_avg = silhouette_score(df4, cluster_labels)
    print("For n_clusters = {}".format(n_clusters) +
          " the avg. silhouette_score is :{}".format(silhouette_avg.round(3)))
```

```python
# Perform PCA using 2 eigenvectors.
n = 2
pca = PCA(n)
Xproj = pca.fit_transform(df4)
eigenvalues = pca.explained_variance_

# Label subway stops by KMean cluster of PCA results.
df4['Cluster'] = KMeans(n_clusters=3, random_state=324).fit_predict(df4)
```

```python
# Create Xproj dataframe and append cluster labels
Xproj_df = pd.DataFrame(Xproj)
Xproj_df['Cluster'] = df4.Cluster

# Group Xproj_df by cluster label to plot with labels.
groups = Xproj_df.groupby('Cluster')

# Plot by clusters
fig, (ax) = pl.subplots(nrows=1, ncols=1, figsize=(16, 4))
colors = ["#9b59b6", "#3498db", "#95a5a6"]
for name, group in groups:
    label = 'Cluster ' + str(name)
    ax.scatter(group[0], group[1], label=label, c=colors[name])
ax.set_xlabel('First Eigenvalue')
ax.set_ylabel('Second Eigenvalue')
ax.set_title('2D PCA Reduction of Subway Stop Ridership Time Series \
Clustered into 3 Groups', fontsize=14)
ax.legend()
ax.text(-2000000, -600000, "Figure 3. 2D PCA reduction of subway stop ridership time series \
clustered into 3 groups. Although, the clustering between groups 0 and 1 does \
\nnot appear meaningful, cluster 2 appears to be unique by comparison. These \
subway stops could be investigated to better understand why \nthey are unique \
from the rest.", fontsize=14)
```
