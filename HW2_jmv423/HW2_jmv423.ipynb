```python
import os
import pandas as pd
import matplotlib.pyplot as pl
%matplotlib inline

print (os.environ['DFDATA'])
```

```python
data = (os.environ['DFDATA'])+'/vum2-yjuc/1414245713/vum2-yjuc'
df = pd.read_csv(filepath_or_buffer = data)
df.head()
```

```python
df = df[[1,10,15]]
df = df.drop([90])
df.head()
```

```python
fig = pl.figure(figsize = (5,5))
ax = fig.add_subplot(111)
ax.plot(range(13),range(13), c='r', label ='No Net Change')
ax.scatter(df[[1]]/1000000,df[[2]]/1000000, c='y', label='NYC Agency')
ax.set_xlim(0,12)
ax.set_ylim(0,12)
ax.legend(loc='upper left', fontsize=9)
ax.grid(True)
ax.set_ylabel('2015 Budget ($ Billions)',fontsize = 9)
ax.set_xlabel('2010 Budget ($ Billions)',fontsize = 9)
ax.set_title('NYC Agency Budget Change (2010 to 2015)', fontsize = 11)
ax.set_aspect(1)
txt = 'Figure 1. New York City agency budget change from 2010 to 2015.\n\
The 6 agencies whose budget exceeded $2 billion are (by descending\n\
2015 budget): the Department of Education, Citywide Pension\n\
Contributions, Miscellaneous, Department of Social Services, Debt\n\
Service, and the Police Department.'
ax.text(0,-3.5, txt, fontsize = 8)
pl.show()
```

```python
data = (os.environ['DFDATA'])+'/zkky-n5j3/1414246141/zkky-n5j3'
df = pd.read_csv(filepath_or_buffer = data)
df.head()
```

```python
df = df[[0,2]]
df.head()
```

```python
df.columns = ['NeversinkDate', 'NeversinkStorage']
x = df['NeversinkDate'][0]
df['NeversinkDate'] = pd.to_datetime(df['NeversinkDate'])
df.head()
```

```python
fig = pl.figure(figsize = (10,4))
ax = fig.add_subplot(111)
ax.plot(df['NeversinkDate'],df['NeversinkStorage']/10000, c='r', label ='Neversink Reservoir')
ax.set_xlim(pd.to_datetime('2008-01-01'),pd.to_datetime('2009-01-01'))
ax.legend(loc='upper right', fontsize=9)
ax.set_ylabel('Volume (10,000,000 US Gal)',fontsize = 9)
ax.set_xlabel('Date (Month)',fontsize = 9)
ax.set_title('Reservoir Storage (01/01/2008 - 01/01/2009)', fontsize = 11)
ax.text(pd.to_datetime('2008-01-01'), 0.75, 'Figure 2. Reservoir storage (01/01/2008 - 01/01/2009). As illustrated, the Neversink Reservoir exhibited \n\
seasonal variability, decreasing in storage during the hot summer months before being replenished in the fall\n\
& the winter until returning to its capacity (approx. 35,000,000 US Gal).')
pl.show()
```
